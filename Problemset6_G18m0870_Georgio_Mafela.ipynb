### Question 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties:

X label name = Month Number

Y label name = Total profit

Title label name = Total profit per month

The line plot graph should look like this:
<img src="Q1.png">

```python
# Answer here
import pandas as pd
import matplotlib.pyplot as plt 

link ="https://raw.githubusercontent.com/GEORGIO-MAF/Data-Visualization/master/company_sales_data.csv" #uploaded my csv file on my github accout and copied the link from there
df = pd.read_csv(link)

profitList=df['total_profit'].tolist()
no_of_month = df['month_number'].tolist()


fig = plt.figure()
fig.suptitle('Total profit per month')

plt.plot(no_of_month,profitList)
plt.xlabel('month number')
plt.ylabel('total profit')
```

### Question 2: Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

Line Style dotted and Line-color should be red

Show legend at the lower right location.

X label name = Month Number

Y label name = Sold units number

Add a circle marker.

Line marker color as read

Line width should be 3

The line plot graph should look like this:
<img src="Q2.png">

```python

fig = plt.figure()

fig.suptitle('Company sales per month')
plt.plot(no_of_month,profitList,color="red",linestyle="--",dashes=(7, 1))
plt.legend(["profit data of last year"],loc="lower right")
plt.xlabel('month number')
plt.ylabel('sold units number')
plt.scatter(no_of_month, profitList, marker="o",c="black")
```

### Question 3: Read all product sales data and show it  using a multiline plot

<img src="Q3.png">
